## Calculating actions for a simulation snapshot using [`pynbody`](http://pynbody.github.io) and [`galpy`](https://github.com/jobovy/galpy)

```python
from snapshot_potential import SnapshotPotential
from galpy.potential import interpRZPotential
from galpy.actionAngle import actionAngleStaeckel, actionAngleAdiabatic
import numpy as np
```

```python
s = pynbody.load('/Users/rokstar/Nbody/runs/12M_hr_diff_coeff0.05/12M_hr_diff_coeff0.05.01000')
pynbody.analysis.angmom.faceon(s,mode='ssc')
```

### Setting up the potentials

First, generate the base `SnapshotPotential` object that performs a calculation of the potential and forces. We will feed this to the `interpRZPotential` to generate a grid on which to interpolate:

```python
snap_pot = SnapshotPotential(s)
```

Now, we generate the `interpRZPotential` which just generates a grid and calls the `snap_pot` to provide the values of the potential at each point:

```python
interp_pot = interpRZPotential(snap_pot,rgrid = (numpy.log(0.01),numpy.log(20.),11), zgrid=(0.,1.0,11),logR=True,interpepifreq=False,interpverticalfreq=False,interpvcirc=False,use_c=False,enable_c=True,interpPot=True,interpDens=False, interpdvcircdr=False)
```

```python
interp_pot.plot()
```

### Getting the actions

```python
aAS= actionAngleStaeckel(pot=interp_pot,delta=0.45,c=True)
```

select just particles in a certain radial region:

```python
sn = pynbody.filt.BandPass('rxy','7 kpc','9 kpc')
```

the action code requires contiguous arrays so we need to make sure we have them 

```python
rxy,vr,vt,z,vz = [np.ascontiguousarray(s.s[sn][x]) for x in ('rxy','vr','vt','z','vz')]
```

```python
jr,lz,jz= aAS(rxy,vr,vt,z,vz)
```

now we make some plots:

```python
h,x,y,im = hist2d(lz/250.,jr/250.,range=[[3,11],[0,1.]],bins=50,cmap=cm.Blues_r)
xlabel('$L_z$ [250 kpc km s$^{-1}$]')
ylabel('$J_R$ [250 kpc km s$^{-1}$]')
xs = (x[:-1]+x[1:])/2.0
ys = (y[:-1]+y[1:])/2.0
contour(xs,ys,log10(h.T),colors='k')
```

```python
h,x,y,im = hist2d(jr/250.,jz/250.,range=[[0,1.0],[0,.1]],bins=50,cmap=cm.Blues_r)
xlabel('$J_R$ [250 kpc km s$^{-1}$]')
ylabel('$J_z$ [250 kpc km s$^{-1}$]')
xs = (x[:-1]+x[1:])/2.0
ys = (y[:-1]+y[1:])/2.0
contour(xs,ys,log10(h.T),colors='k')
```
